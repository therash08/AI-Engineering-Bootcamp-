# Understanding Numerical and Categorical Columns


## 📊 Understanding Numerical Columns

Numerical columns contain **numbers** (int, float) and are used for calculations.

---

#### 🔹 Basic Summary

```python
# summary statistics for numerical columns
df.describe()
```

**Includes:**

* count
* mean
* std (standard deviation)
* min
* 25%, 50%, 75% (quartiles)
* max

---

#### 🔹 Key Metrics

```python
# average value
df["Age"].mean()

# middle value
df["Age"].median()

# most frequent value
df["Age"].mode()

# minimum value
df["Age"].min()

# maximum value
df["Age"].max()

# index of max value
df["Age"].idxmax()

# spread of data
df["Age"].std()
```

---

#### 🔹 When to Use What

* **mean** → when data is **not skewed**
* **median** → when data has **outliers/skew**
* **std** → understand variation/spread
* **min/max** → range of values

---

#### 🔹 Quick Insight Checks

```python
# check skewness
df["Age"].skew()

# check distribution
df["Age"].hist()
```

---

## 📊 Understanding Categorical Columns

Categorical columns contain **labels / categories** (text or discrete values).

---

#### 🔹 Basic Summary

```python
# summary for categorical columns
df.describe(include="object")
```

**Includes:**

* count
* unique
* top (most frequent value)
* freq (frequency of top value)

---

#### 🔹 Key Operations

```python
# most frequent value (mode)
df["Gender"].mode()

# count of each category
df["Gender"].value_counts()

# unique values
df["Gender"].unique()

# number of unique categories
df["Gender"].nunique()
```

---

#### 🔹 Proportion / Percentage

```python
# percentage distribution
df["Gender"].value_counts(normalize=True)
```

---

#### 🔹 When to Use What

* **mode** → most common category
* **value_counts()** → full distribution
* **nunique()** → number of categories
* **normalize=True** → percentage view

---

## 🔥 Key Difference

| Feature        | Numerical Columns | Categorical Columns          |
| -------------- | ----------------- | ---------------------------- |
| Data Type      | int, float        | object, category             |
| Examples       | Age, Salary, Fare | Gender, City, Embarked       |
| Key Metrics    | mean, median, std | mode, frequency              |

---

## 🚀 Summary

* Numerical → use **mean, median, std** for analysis
* Categorical → use **mode, value_counts**
* Always check distribution before deciding method
* Median is safer when data is skewed



# 📊 Mastering Data Selection in Pandas

This note covers essential techniques for **selecting, indexing, slicing, and filtering data** in Pandas using `df[]`, `df.loc[]`, `df.iloc[]`.

---

## 🔹 1. `df[]` → Basic Column Selection

#### ✅ Select a single column

```python
df["Age"]
```

#### ✅ Select multiple columns

```python
df[["Age", "Fare"]]
```

#### ✅ Boolean filtering (very common)

```python
df[df["Age"] > 30]
```

👉 Think: `df[]` = **columns + filtering**

---

## 🔹 2. `df.loc` → Label-based selection

Uses **row labels (index names)** and **column names**

#### ✅ Syntax

```python
df.loc[row_labels, column_labels]
```

#### ✅ Select specific row

```python
df.loc[0]
```

#### ✅ Select multiple rows

```python
df.loc[[0, 1, 2]]
```

#### ✅ Select rows + columns

```python
df.loc[0:5, ["Age", "Fare"]]
```

#### ✅ Conditional selection

```python
df.loc[df["Age"] > 30, ["Age", "Pclass"]]
```

👉 Important:

* Slicing is **inclusive** (`0:5` includes 5)

---

## 🔹 3. `df.iloc` → Position-based selection

Uses **integer positions (index positions)**

#### ✅ Syntax

```python
df.iloc[row_index, column_index]
```

#### ✅ Select first row

```python
df.iloc[0]
```

#### ✅ Select multiple rows

```python
df.iloc[0:3]
```

#### ✅ Select rows + columns

```python
df.iloc[0:5, 0:2]
```

👉 Important:

* Slicing is **exclusive** (`0:5` → stops before 5)

---

## 🔥 Key Differences (Very Important)

| Feature        | `df.loc`    | `df.iloc`      |
| -------------- | ----------- | -------------- |
| Type           | Label-based | Position-based |
| Index used     | Names       | Integer index  |
| Slice behavior | Inclusive   | Exclusive      |

---

## 🔹 Examples Together

```python
# First 5 rows (position-based)
df.iloc[:5]

# First 5 rows (label-based)
df.loc[:4]

# Select specific value
df.loc[0, "Age"]
df.iloc[0, 2]

# select first 5 rows of first 2 columns
df.iloc[0:5, 0:2]

```

---

#### 🔹 When to Use What

* Use `df[]` → simple column selection or filtering
* Use `df.loc` → when working with **column names / labels / conditions**
* Use `df.iloc` → when working with **row/column positions (numbers)**

---

#### 🔹 Quick Memory Trick

* `loc` → **L**abel
* `iloc` → **I**nteger



# 🔍 Boolean / Conditional Filtering in Pandas

Boolean filtering is used to **select rows based on conditions**. It is one of the most important operations in EDA and data preprocessing.

---

## 🔹 1. Basic Filtering

```python
# select rows where age is greater than 30
df[df["Age"] > 30]
````

---

## 🔹 2. Filtering with Specific Columns

```python
# select name, age and fare for passengers whose age is greater than 30
df.loc[df["Age"] > 30, ["Name", "Age", "Fare"]]
```

---

## 🔹 3. Multiple Conditions (AND / OR)

```python
# select passengers whose age > 30 AND fare > 50
df[(df["Age"] > 30) & (df["Fare"] > 50)]

# select passengers whose age > 30 OR fare > 50
df[(df["Age"] > 30) | (df["Fare"] > 50)]
```

---

## 🔹 4. NOT Condition

```python
# select passengers whose age is NOT greater than 30
df[~(df["Age"] > 30)]

# select passengers whose not in first class
df[~(df["Pclass"] == 1)]
```

---

## 🔹 5. Using isin()

```python
# select passengers who embarked from 'C' or 'S'
df[df["Embarked"].isin(["C", "S"])]

# select passengers who does not embarked from 'C' or 'S'
df[~df["Embarked"].isin(["C", "S"])]
```

---

## 🔹 6. Range Filtering

```python
# select passengers whose age is between 20 and 30 (inclusive)
df[df["Age"].between(20, 30)]
```

---

## 🔹 7. Missing Value Filtering

```python
# select rows where age is missing
df[df["Age"].isnull()]

# select rows where age is NOT missing
df[df["Age"].notnull()]
```

---

## 🔹 8. String-based Filtering

```python
# select rows where name contains 'John'
df[df["Name"].str.contains("John")]

# select rows where name starts with 'A'
df[df["Name"].str.startswith("A")]

# select rows where name ends with 'son'
df[df["Name"].str.endswith("son")]
```

---

## 🔹 9. Query Method (Readable Alternative)

```python
# select passengers whose age > 30 and fare > 50
df.query("Age > 30 and Fare > 50")
```

---

## 🔹 10. Conditional Assignment

```python
# create/update 'Category' column as 'Child' where age < 18
df.loc[df["Age"] < 18, "Category"] = "Child"
```

---

## 🔹 11. Counting with Condition

```python
# count number of passengers whose age > 30
(df["Age"] > 30).sum()
```

---

## ⚠️ Important Rules

```python
# ❌ WRONG (will throw error)
df[df["Age"] > 30 and df["Fare"] > 50]

# ✅ CORRECT (use & and parentheses)
df[(df["Age"] > 30) & (df["Fare"] > 50)]
```

---

## 🚀 Summary

* Use `df[]` for simple filtering
* Use `df.loc` for filtering + column selection
* Use `&`, `|`, `~` instead of `and`, `or`, `not`
* Always wrap conditions in parentheses
* Handle missing values separately with `isnull()` / `notnull()`


# 📊 GroupBy and Sorting in Pandas

Used for:

* **Grouping data by categories**
* **Applying aggregations (mean, sum, count, etc.)**
* **Sorting results for better insights**

---

## 🔹 1. GroupBy Basics

```python
# group data by a column
df.groupby("Gender")
```

👉 Creates a grouped object (no computation yet)

---

## 🔹 2. Aggregation Functions

```python
# average age per gender
df.groupby("Gender")["Age"].mean()

# total fare per gender
df.groupby("Gender")["Fare"].sum()

# count of records per gender
df.groupby("Gender")["Age"].count()
```

---

## 🔹 3. Multiple Aggregations

```python
# multiple stats on Age column
df.groupby("Gender")["Age"].agg(["mean", "min", "max"])
```

---

## 🔹 4. Grouping by Multiple Columns

```python
# group by Gender and Class
df.groupby(["Gender", "Pclass"])["Fare"].mean()
```

---

## 🔹 5. Reset Index (Very Common)

```python
# convert grouped result to normal dataframe
df.groupby("Gender")["Age"].mean().reset_index()
```

---

## 🔹 6. Named Aggregations (Clean Output)

```python
# custom column names
df.groupby("Gender").agg(
    avg_age=("Age", "mean"),
    max_fare=("Fare", "max")
)
```

---

## 🔹 7. Filtering Groups

```python
# keep groups where average age > 30
df.groupby("Gender").filter(lambda x: x["Age"].mean() > 30)
```

---

## 🔹 8. Transform (Very Important)

```python
# create new column: mean age per group
df["mean_age_by_gender"] = df.groupby("Gender")["Age"].transform("mean")
```

👉 Keeps original shape (unlike agg)

---

## 🔹 9. Sorting Basics

#### Sort by Column Values

```python
# sort by Age (ascending)
df.sort_values(by="Age")

# sort by Age (descending)
df.sort_values(by="Age", ascending=False)
```

---

#### Sort by Multiple Columns

```python
# sort by Gender then Age
df.sort_values(by=["Gender", "Age"])
```

---

#### Sort Index

```python
# sort by index
df.sort_index()
```

---

## 🔹 10. Sorting Grouped Results

```python
# average fare per class, sorted descending
df.groupby("Pclass")["Fare"].mean().sort_values(ascending=False)
```

---

## 🔹 11. Top / Bottom Values

```python
# top 5 highest fares
df.nlargest(5, "Fare")

# lowest 5 ages
df.nsmallest(5, "Age")
```

---

## 🔹 12. Common Patterns (Real EDA)

```python
# average fare by class (sorted)
df.groupby("Pclass")["Fare"].mean().sort_values(ascending=False)

# count of passengers per gender
df["Gender"].value_counts()

# average age by gender and class
df.groupby(["Gender", "Pclass"])["Age"].mean().reset_index()
```

---

## 🔥 Key Concepts

* `groupby()` → split data into groups
* aggregation (`mean`, `sum`, etc.) → apply function
* `transform()` → keeps original shape
* `sort_values()` → organize results

---

## ⚠️ Common Mistakes

```python
# ❌ forgetting aggregation
df.groupby("Gender")   # no output

# ❌ wrong chaining order
df.sort_values().groupby("Gender")  # usually not useful
```

---

## 🚀 Summary

* Use **groupby + aggregation** to summarize data
* Use **sorting** to rank and analyze patterns
* Use **transform** when you need group info per row



# 📊 Handling Missing Values

## 🔹 Step 1: Check Missing Values

```python
df.isnull().sum()              # count
df.isnull().mean() * 100       # percentage
```

---

## 🔹 Step 2: Decide Strategy Based on Missing %

| Missing % | Action               |
| --------- | -------------------- |
| < 5%      | Drop rows            |
| 5% – 60%  | Impute (fill values) |
| > 60–70%  | Drop column          |


#### ⚠️ N.B. (Important)

👉 The **5%, 60%, 70% thresholds are NOT fixed rules** — they are just general guidelines.

* In real projects, the decision depends on:

  * 📊 Dataset size
  * 🎯 Importance of the feature
  * 🧠 Domain knowledge
  * 📉 Impact on model performance

👉 Example:

* A column with **80% missing** might still be useful → you may **keep it**
* A column with **2% missing** but very important → you should **not drop rows blindly**


---

## 🔹 Step 3: Analyze Column Type

-  🔢 Numerical Columns

- 🔤 Categorical Columns

⚠️ N.B. (Important):
Some columns may look numerical but are actually categorical in nature.
Example: `df["Pclass"] = 1, 2, 3` → represents passenger classes, not real numeric values.

👉 So always check meaning of values, not just data type.

---

## 🔹 Step 4: For Numerical Columns → Check Distribution

### ✅ Check Skewness

```python
df['Age'].skew()
```

### 📌 Interpretation

| Skewness     | Meaning      | Action         |
| ------------ | ------------ | -------------- |
| -0.5 to +0.5 | Normal       | Use **Mean**   |
| > +0.5       | Right skewed | Use **Median** |
| < -0.5       | Left skewed  | Use **Median** |

👉 Key idea:

* **Mean → sensitive to outliers**
* **Median → robust (safer choice)**

👉 The ±0.5 skewness rule is not strict, but a common guideline used in practice.
Always confirm with visualization (histogram/boxplot) and domain understanding before choosing mean or median.

---

### 📊 Optional Visualization (very helpful)

```python
df['Age'].hist()
```

```python
import seaborn as sns
sns.boxplot(x=df['Age'])
```

---

## 🔹 Step 5: Apply Imputation

### 🔢 Numerical

#### Mean (Normal Data)

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

#### Median (Skewed Data)

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

---

### 🔤 Categorical

#### Mode

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

---

## 🔹 Step 6: Alternative Options

#### ✅ Drop Column

```python
df.drop(columns=['Cabin'], inplace=True)
```

Use when:

* Too many missing values (>70%)
* Not important

---

#### ✅ Drop Rows

```python
df.dropna(inplace=True)
```

Use when:

* Very few missing rows (<5%)

---


## ⚡ Final Decision Flow (Super Important)

1. Check missing %
2. Identify column type
3. For numerical → check skewness
4. Choose:

   * Mean (normal)
   * Median (skewed)
   * Mode (categorical)
5. Consider advanced methods if needed





# 📌 Pandas: `inplace=True` vs Reassignment

In Pandas, many operations can either:

* **modify the original DataFrame**, or
* **return a new modified DataFrame**

---

## 🔹 1. Using `inplace=True`

```python
# drop column and modify original df directly
df.drop("Age", axis=1, inplace=True)
```

#### ✅ What happens?

* Changes are applied **directly to `df`**
* No need to assign back

---

## 🔹 2. Without `inplace` (Default behavior)

```python
# drop column but NOT modifying original df
df.drop("Age", axis=1)
```

#### ⚠️ What happens?

* Returns a **new DataFrame**
* Original `df` remains unchanged

---

# 🔹 3. Reassignment (Recommended way)

```python
# explicitly reassign the result
df = df.drop("Age", axis=1)
```

#### ✅ What happens?

* New DataFrame is created
* Assigned back to `df`

---

#### 🔥 Key Difference

| Approach       | Modifies Original?    | Returns New Object? |
| -------------- | --------------------- | ------------------- |
| `inplace=True` | ✅ Yes                 | ❌ No                |
| No inplace     | ❌ No                  | ✅ Yes               |
| Reassignment   | ✅ Yes (via overwrite) | ✅ Yes               |

---

## 🔹 4. Example Comparison

```python
# inplace version
df.drop("Age", axis=1, inplace=True)

# reassignment version (preferred)
df = df.drop("Age", axis=1)
```

👉 Both achieve the same final result

---

## 🔹 5. Why Reassignment is Preferred 🚀

Modern Pandas best practice:

```python
df = df.drop(...)
```

#### Reasons:

* More **readable & explicit**
* Easier to **debug**
* Works better with **method chaining**
* `inplace=True` is being discouraged in many cases

---

## 🔹 6. Method Chaining Example

```python
df = (
    df
    .drop("Age", axis=1)
    .rename(columns={"Fare": "Ticket_Fare"})
)
```

👉 Not possible with `inplace=True`

---

## 🔹 7. Common Functions Supporting `inplace`

```python
df.drop(...)
df.rename(...)
df.fillna(...)
df.sort_values(...)
```

---

## ⚠️ Common Mistake

```python
# ❌ WRONG (df becomes None)
df = df.drop("Age", axis=1, inplace=True)
```

👉 Why?

* `inplace=True` returns **None**

---

## ✅ Correct Ways

```python
# option 1
df.drop("Age", axis=1, inplace=True)

# option 2 (recommended)
df = df.drop("Age", axis=1)
```

---

## 🚀 Summary

* `inplace=True` → modifies original DataFrame
* No `inplace` → returns new DataFrame
* Best practice → use reassignment (`df = df...`)
* Never combine `inplace=True` with assignment

#### 1. Countplot with Hue (MOST USED)

```python
sns.countplot(x="gender", hue="survived", data=df)
```

✔ Use when:

* You want relationship + frequency together
* Very intuitive visualization

---

#### 2. Crosstab (table view)

```python
pd.crosstab(df["gender"], df["survived"])
```

✔ Use when:

* You want exact counts

---

#### 3. Percentage view (very useful)

```python
pd.crosstab(df["gender"], df["survived"], normalize="index") * 100
```

✔ Use when:

* You want proportions instead of raw numbers

---

## F. Correlation between multiple features

**Heatmap**
```python
sns.heatmap(df.corr(), annot=True)
```

✔ Use when:

* Feature selection in ML
* Finding correlated variables

## 🧠 Final Cheat Sheet (Super Important)


| Type of Data                | Best Plot (Subjective)   |
| ----------------------------| ------------------------ |
| Numerical (1 variable)      | Histogram / Boxplot      |
| Categorical (1 variable)    | Countplot                |
| Numerical vs Numerical      | Scatter plot             |
| Time trend                  | Line plot                |
| Numerical vs Categorical    | Boxplot / Violin         |
| Categorical vs Categorical  | Countplot (hue)          |
| Feature correlation         | Heatmap                  |

---

## ⚡ Golden Rule (EDA Thinking)

* 📊 Distribution → Histogram / KDE
* 📦 Outliers → Boxplot
* 🔵 Relationship → Scatter
* 📈 Trend → Line
* 📦 Group comparison → Boxplot
* 🔗 Category relationship → Countplot (hue)
